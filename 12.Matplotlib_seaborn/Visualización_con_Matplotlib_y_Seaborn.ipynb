### Múltiples funciones

```python
# Datos
x = np.linspace(0, 2 * np.pi)
sin = np.sin(x)
cos = np.cos(x)

# Figura y Marco
fig, ax = plt.subplots()

#Gráficos
ax.plot(x, sin)
ax.plot(x, cos);
```

```python
# Añadir una leyenda
fig, ax = plt.subplots()
ax.plot(x, sin, label='sin')
ax.plot(x, cos, label='cos')
ax.legend()
fig;
```

```python
# Ecuación de Latex en elementos textuales
fig, ax = plt.subplots()
ax.plot(x, sin, label='$f_1(x) = sin(x)$')
ax.plot(x, cos, label='$f_2(x) = cos(x)$')
ax.legend()
fig;
```

```python
# Ubicación de la leyenda
ax.legend(loc='center right')
fig
```

### Estilos

```python
fig, ax = plt.subplots()

sin_style = dict(linewidth=5, color='darkorange')
cos_style = dict(marker='o', markerfacecolor='limegreen', color='darkgreen')

ax.plot(x, sin, label='$f_1(x) = sin(x)$', **sin_style)
ax.plot(x, cos, label='$f_2(x) = cos(x)$', **cos_style)

ax.legend()

fig;
```

### Acotar los ejes

```python
ax.set_xlim(0, np.pi / 2)
ax.set_ylim(0, 1)
fig
```

### Anotaciones

#### Vamos a anotar los puntos de corte donde sin(x)=cos(x). Vamos a poner también una flecha con una etiqueta, las coordenadas de solución y la proyección del punto hacia ambos ejes


```python
# Calcular el punto de corte 
xsol = np.pi / 4 + np.pi * 0
ysol = np.sin(xsol)
xsol, ysol
```

```python
# Figura
fig, ax = plt.subplots()
ax.plot(x, sin, label='$f_1(x) = sin(x)$')
ax.plot(x, cos, label='$f_2(x) = cos(x)$')
ax.set_xlim(0, np.pi / 2)
ax.set_ylim(0, 1)

# Flecha
ax.annotate('$sin(x) = cos(x)$',
            xy=(xsol, ysol),
            xytext=(1.2, 0.8),
            arrowprops=dict(facecolor='black', shrink=0.05))

# Texto indicando el punto de corte '(0.79, 0.71)':
ax.text(0.5, 0.75, '(0.79, 0.71)')

ax.plot([xsol, xsol], [0, ysol], color='gray', linestyle='--')
ax.plot([0, xsol], [ysol, ysol], color='gray', linestyle='--')
```

## Gráfico de barras e Histograma

#### Gráfico de barras

```python
# Datos
x = 0.5 + np.arange(5)
y = np.random.uniform(2, 7, len(x))

# Figura
fig, ax = plt.subplots(figsize=(8, 5))

# Gráfico de barras
ax.bar(x, y, width=1,edgecolor="white", linewidth=0.7)   

print('x:', x)
print('y:', y)
plt.show()
```

```python
# Título
ax.set_title('Gráfico de barras')

# Etiqueta en cada barra
xlabels=['1ra categoria', '2da categoria','3ra categoria','4ta categoria','5ta categoria']
ax.set_xticks(x)
ax.set_xticklabels(xlabels, rotation=90)
fig
```

#### Histograma

```python
# Datos
x = 4 + np.random.normal(0, 1.5, 200)

# Figura
fig, ax = plt.subplots()

# Histograma
ax.hist(x, bins=8, linewidth=0.5, edgecolor="white")

plt.show()

```

```python
# Modificando el estilo:

# bins: intervalos de agrupación
# rwidth: ancho de cada barra
# color: color
# alpha: transparencia

fig, ax = plt.subplots(figsize=(15, 8))
ax.hist(x, bins=14, rwidth=0.95, color='deeppink', alpha=0.2)
plt.show()
```

## Histogramas múltiples

```python
# Datos
n = 5000

mean_mu1 = 60
sd_sigma1 = 15
data1 = np.random.normal(mean_mu1, sd_sigma1, n)

mean_mu2 = 80
sd_sigma2 = 15
data2 = np.random.normal(mean_mu2, sd_sigma2, n)
```

```python
plt.figure(figsize=(8,6))

plt.hist(data1, bins=100, alpha=0.5, label="data1")
plt.hist(data2, bins=100, alpha=0.5, label="data2")

plt.xlabel("Data", size=14)
plt.ylabel("Count", size=14)
plt.title("Multiple Histograms with Matplotlib")
plt.legend(loc='upper right')

plt.show()
```

## Boxplots múltiples 

```python
# Datos
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))
np.shape(D)
```

```python
# Boxplots múltiples con estilo por defecto
fig, ax = plt.subplots(figsize=(10, 8))
ax.boxplot(D, positions=[2, 4, 6], widths=1.5)
plt.show()
```

```python
# Modifcando el estilo

#positions: posición de cada boxplot
#widths: ancho de la caja
#patch_artist: modificar estilo
#showmeans: señalar la media con un símbolo
#showfliers: mostrar outliers o no
#flierprops: estilo de los outliers
#medianprops: estilo de la mediana
#boxprops: estilo de las cajas
#whiskerprops: estilo de los bigotes
#capprops: estilo de los límites del boxplot


fig, ax = plt.subplots(figsize=(10, 8))
ax.boxplot(D, positions=[2, 4, 6], widths=1.5, patch_artist=True,
                showmeans=False, showfliers=True,
                flierprops={"color": "black", "markerfacecolor": "cyan", "markersize": 9},
                medianprops={"color": "white", "linewidth": 1.5},
                boxprops={"facecolor": "pink", "edgecolor": "white", "linewidth": 0.5},
                whiskerprops={"color": "magenta", "linewidth": 2.5},
                capprops={"color": "green", "linewidth": 3.5})
plt.show()
```

## Gráfico de dispersión

```python
# Datos
n = 200

mean_mu1 = 60
sd_sigma1 = 15
x = np.random.normal(mean_mu1, sd_sigma1, n)

y = 2*x + np.random.normal(5, 4, n)
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
ax.scatter(x, y, 
               s=40, # tamaño de los puntos
               color='green', # color
               alpha=0.7, #transparencia de los puntos
               edgecolors='none')

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.title("Scatterplot with Matplotlib")

plt.show()
```

## Mapa de calor para matriz de correlaciones


```python
# Datos
a=np.array([[1, 1, 2, 2, 3],  
            [2, 2, 3, 3, 5],  
            [1, 4, 2, 2, 3]]) 
a
```

```python
mcor=np.corrcoef(a)
mcor
```

```python
fig, ax = plt.subplots()
im=ax.imshow(mcor,cmap='Reds')
fig.colorbar(im, ax=ax, label='Correlaciones')
plt.show()
```

### Scatterplot 

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.set_theme(style="darkgrid")


g = sns.jointplot(x="total_bill", y="tip", data=tips,
                  kind="scatter",  #kind={ “scatter” | “kde” | “hist” | “hex” | “reg” | “resid” }
                  xlim=(0, 60), ylim=(0, 12),
                  color="m", height=7)
```

### Boxplot múltiple


```python
# Figura (escala logarítmica)
f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale("log")

# Dataset 'planets'
planets = sns.load_dataset("planets")

# Boxplot
sns.boxplot(x="distance", y="method", data=planets,
            whis=[0, 100], width=.6, palette="vlag")

# Añadir puntos de datos
sns.stripplot(x="distance", y="method", data=planets,
              size=4, color=".3", linewidth=0)

# Configuración visual
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```
