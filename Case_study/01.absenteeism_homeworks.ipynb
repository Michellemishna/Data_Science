```python
column_names = ['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Daily Work Load Average', 'Body Mass Index',
       'Education', 'Children', 'Pets', 27,
       28, 29, 30, 31, 32, 33, 34, 36, 37, 38, 39, 40, 41, 43, 46, 47, 48,
       49, 50, 58, 'Absenteeism Time in Hours']
```

```python
type(df_checkpoint['Date'][0])
```

```python
df_checkpoint['Date'] = pd.to_datetime(df_checkpoint['Date'], dayfirst=True)
df_checkpoint['Date']
```

```python
list_months = []
```

```python
for i in range(df_checkpoint.shape[0]):
    list_months.append(df_checkpoint['Date'][i].month)
    
```

```python
list_months
```

```python
len(list_months)
```

```python
df_checkpoint['Month Value'] = list_months
df_checkpoint.head()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_checkpoint['Day of the Week'] = df_checkpoint['Date'].apply(date_to_weekday)
df_checkpoint.head()
```
